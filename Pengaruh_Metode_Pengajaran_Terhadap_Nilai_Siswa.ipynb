## Uji Normalitas Multivariat

Keterangan:


* DI = Differentiated Instruction(Pengajaran Berdiferensiasi) = Metode pengajaran yang disesuaikan dengan kebutuhan, minat, gaya belajar, dan tingkat kemampuan masing-masing siswa.

* LBI = Lecture-based Instruction(Pengajaran Berbasis Ceramah) = Metode pengajaran tradisional di mana guru aktif menjelaskan materi secara lisan, sementara siswa mendengarkan dan mencatat.

* TBL = Technology-based Learning(Pembelajaran Berbasis Teknologi) = Metode pengajaran yang memanfaatkan alat dan platform teknologi (seperti komputer, internet, aplikasi belajar, video interaktif) untuk mendukung proses belajar.




Hipotesis:


* H₀: Data berdistribusi normal multivariat.
* H₁: Data tidak berdistribusi normal multivariat.



```python
data_DI = data_manova2.loc[data_manova2['TeachingMethod']=='Differentiated Instruction', ['EnglishScore',	'MathScore',	'ChemistryScore',	'PhysicsScore',	'BiologyScore']]
data_LBI = data_manova2.loc[data_manova2['TeachingMethod']=='Lecture-based Instruction', ['EnglishScore',	'MathScore',	'ChemistryScore',	'PhysicsScore',	'BiologyScore']]
data_TBL = data_manova2.loc[data_manova2['TeachingMethod']=='Technology-based Learning', ['EnglishScore',	'MathScore',	'ChemistryScore',	'PhysicsScore',	'BiologyScore']]
```

```python
print("uji normalitas multivariat Differentiated Instruction: ", pg.multivariate_normality(data_DI, alpha=0.05))

print("uji normalitas multivariat Lecture-based Instruction: ", pg.multivariate_normality(data_LBI, alpha=0.05))

print("ujl normalitas multivariat Technology-based Learning: ", pg.multivariate_normality(data_TBL, alpha=0.05))
```

karena p-value < 0.05 maka Data tidak berdistribusi normal multivariat, sehingga kita perlu menerapkan transformasi data

```python
#Transformasi
data_DI_trans = np.log(data_DI)
data_LBI_trans = np.log(data_LBI)
data_TBL_trans = np.log(data_TBL)
```

Lalu lakukan lagi uji normalitas multivariatnya

```python
print("uji normalitas multivariat Differentiated Instruction: ", pg.multivariate_normality(data_DI_trans, alpha=0.05))

print("uji normalitas multivariat Lecture-based Instruction: ", pg.multivariate_normality(data_LBI_trans, alpha=0.05))

print("ujl normalitas multivariat Technology-based Learning: ", pg.multivariate_normality(data_TBL_trans, alpha=0.05))
```

karena p-value > 0.05 maka Data berdistribusi normal multivariat

## Uji Homogenitas

```python
#Gabungkan lagi datanya
data_trans_manova2 = pd.concat([data_DI_trans, data_LBI_trans, data_TBL_trans])
data_trans_manova2['TeachingMethod'] = data_manova2['TeachingMethod'].copy()
data_trans_manova2
```

Hipotesisi:


* H₀: Hubungan antar variabel data homogen.
* H₁: Hubungan antar variabel data  tidak homogen.



```python
#Uji Homogenitas
pg.box_m(data_trans_manova2, dvs=['EnglishScore',	'MathScore',	'ChemistryScore',	'PhysicsScore',	'BiologyScore'], group='TeachingMethod')
```

karena p-value > 0.05 maka hubungan antar variabel (varians dan kovarians) homogen


## MANOVA

```python
manova_model = MANOVA.from_formula('EnglishScore + MathScore + ChemistryScore + PhysicsScore + BiologyScore ~ TeachingMethod', data=data_trans_manova2)
manova_results = manova_model.mv_test()
print(manova_results)
```

Semua statistik uji Manova menunjukkan p-value < 0.05, maka terdapat perbedaan yang signifikan dalam gabungan nilai mata pelajaran (English, Math, Chemistry, Physics, Biology) berdasarkan metode pengajaran.

## Uji Post Hoc

Disini kita akan melihat pasangan kelompok mana yang berbeda signifikan antar ke-3nya

```python
tukey_EnglishScore = pg.pairwise_tukey(data=data_trans_manova2, dv='EnglishScore', between='TeachingMethod')
print("Tukey Post-hoc untuk variabel EnglishScore")
tukey_EnglishScore
```

```python
tukey_MathScore = pg.pairwise_tukey(data=data_trans_manova2, dv='MathScore', between='TeachingMethod')
print("Tukey Post-hoc untuk variabel MathScore")
tukey_MathScore
```

```python
tukey_ChemistryScore = pg.pairwise_tukey(data=data_trans_manova2, dv='ChemistryScore', between='TeachingMethod')
print("Tukey Post-hoc untuk variabel ChemistryScore")
tukey_ChemistryScore
```

```python
tukey_PhysicsScore = pg.pairwise_tukey(data=data_trans_manova2, dv='PhysicsScore', between='TeachingMethod')
print("Tukey Post-hoc untuk variabel PhysicsScore")
tukey_PhysicsScore
```

```python
tukey_BiologyScore = pg.pairwise_tukey(data=data_trans_manova2, dv='BiologyScore', between='TeachingMethod')
print("Tukey Post-hoc untuk variabel BiologyScore")
tukey_BiologyScore
```

Ternyata semuanya berbeda signifikan, sekarang mari kita lihat visualisasi datanya

## Visualisasi Data
